# Importing BeautifulSoup for Web Scraping

```python
from bs4 import BeautifulSoup
import requests
```

# Declaring User Agent of the system and browser

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.182 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})
```

# URL of the website

```python
URL = "https://www.amazon.in/s?k=mobile+phones+under+20000&rh=p_36%3A-2000000&qid=1616434552&rnid=1318502031&ref=sr_nr_p_36_5"
```

```python
page = requests.get(URL, headers=HEADERS) #Sending Request
page
```

Request Approved

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
Mobile_names=soup.find_all('a',class_="a-link-normal a-text-normal")
Mobile_names
```

```python
Mobile_names_list=[]
for i in Mobile_names:
    Mobile_names_list.append(i.text.replace('\n',''))   #replacing unwanted '\n' containing in each text when it extracted

Mobile_names_list
```

```python
Mobile_names_list_1=Mobile_names_list
Mobile_names_list_1
```

```python
for i in range(len(Mobile_names_list)):
    Mobile_names_list[i]=Mobile_names_list[i].split("(")[0]

Mobile_names_list
```

```python
Features=soup.find_all('a',class_="a-link-normal a-text-normal")
Features
```

```python
Features_list=[]
for i in Features:
    Features_list.append(i.text.replace('\n',''))   #replacing unwanted '\n' containing in each text when it extracted

Features_list
```

```python
for i in range(len(Features_list)):
    Features_list[i]=Features_list[i].split("(")[1]

Features_list
```

```python
Price=soup.find_all('span',class_="a-price-whole")
Price
```

```python
Price_list=[]
for i in Price:
    Price_list.append(i.text.replace('\n',''))   #replacing unwanted '\n' containing in each text when it extracted

Price_list
```

```python
Image_URLs=soup.find_all('img',class_="s-image")
Image_URLs
```

```python
Image_URLs_list=[]
for i in Image_URLs:
    Image_URLs_list.append(i.get("srcset"))   #replacing unwanted '\n' containing in each text when it extracted

Image_URLs_list
```

```python
Rating=soup.find_all('span',class_="a-icon-alt")
Rating
```

```python
Rating_list=[]
for i in Rating:
    Rating_list.append(i.text.replace('\n',''))   #replacing unwanted '\n' containing in each text when it extracted

Rating_list
```

```python
print(len(Mobile_names_list),len(Features_list),len(Price_list),len(Image_URLs_list),len(Rating_list))
```

```python
Rating_list=Rating_list[:20]
```

```python
print(len(Rating_list))
```

```python
#Creating a DataFrame containing Mobiles_Under_20000.
import pandas as pd
Mobiles_Under_20000=pd.DataFrame()
Mobiles_Under_20000['Name']=Mobile_names_list
Mobiles_Under_20000['Features']=Features_list
Mobiles_Under_20000['Price']=Price_list
Mobiles_Under_20000['Image URLs']=Image_URLs_list
Mobiles_Under_20000['Rating']=Rating_list
Mobiles_Under_20000
```
