# Importing BeautifulSoup for Web Scraping

```python
from bs4 import BeautifulSoup
import requests
```

```python
#sending request to website for Web Scraping
page = requests.get("https://forecast.weather.gov/MapClick.php?CityName=City+View&state=SC&site=GSP&textField1=34.862&textField2=-82.4268&e=0#.YFRMR9zhVPY")
page
```

```python
#whole html document
page.content
```

```python
#arranging the html codes in a readable manner 
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
prd = soup.find_all('div', class_ = "col-sm-2 forecast-label")
period = []

for i in prd:
    period.append(i.text.replace("\n",""))
period
```

```python
desc = soup.find_all('div', class_ = "col-sm-10 forecast-text")
description = []

for i in desc:
    description.append(i.text.replace("\n",""))

shrt_desc_1 = [i.split(',')[0]for i in description]
shrt_desc_2 = [i.split(', ')[1]for i in description]
```

```python
shrt_desc_2 =[i.split('.')[0]for i in shrt_desc_2]
```

```python
shrt_desc_1=[i.split('.')[-1]for i in shrt_desc_1]
```

```python
temperature = [i + '°F' for i in shrt_desc_2]
```

```python
shrt_temp = [i.split(' ',3)[-1]for i in temperature]
```

```python
import pandas as pd
SanFrancisco_Forecast = {"Period": period, "Approx_temp": shrt_temp, "Temprature": temperature, 
                         "Short_desc": shrt_desc_1, "Description":description}
San_Francisco_Forecast = pd.DataFrame(SanFrancisco_Forecast)
San_Francisco_Forecast
```
